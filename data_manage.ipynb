```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/response.csv")
```

```python
data
```

```python
def get_avg_pain(row):
    if row["context"] == "Acute Cancer Pain":
        temp = pd.read_csv("data/acute_cancer_responses.csv")
        avg = temp["pain_index"].mean()
        return avg
    elif row["context"] == "Acute Non Cancer Pain":
        temp = pd.read_csv("data/acute_non_cancer_responses.csv")
        avg = temp["pain_index"].mean()
        return avg
    elif row["context"] == "Chonic Cancer Pain":
        temp = pd.read_csv("data/chronic_cancer_responses.csv")
        avg = temp["pain_index"].mean()
        return avg
    elif row["context"] == "Chonic Non Cancer Pain":
        temp = pd.read_csv("data/chronic_non_cancer_responses.csv")
        avg = temp["pain_index"].mean()
        return avg
    elif row["context"] == "Post Op Pain":
        temp = pd.read_csv("data/post_op_responses.csv")
        avg = temp["pain_index"].mean()
        return avg
```

```python
data["pain"] = data.apply(get_avg_pain, axis=1)
```
